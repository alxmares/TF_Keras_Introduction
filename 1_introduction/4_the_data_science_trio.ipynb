# Example with NumPy, Pandas, and Matplotlib

This Jupyter Notebook provides a simple example of using NumPy, Pandas, and Matplotlib. We'll load a dataset, perform some basic manipulation, and visualize the results.


```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading and Preparing the Data

We will use Pandas to load a dataset and NumPy for data manipulation.


```python
# Loading a dataset with Pandas
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset
print(data.head())
```

## Data Manipulation with Pandas

We'll count the number of countries in each region.


```python
# Counting the number of countries per region
countries_per_region = data.groupby('Region').size()

# Displaying the counts
print(countries_per_region)
```

## Visualizing the Data with Matplotlib

Now, let's create a bar chart to visualize the number of countries in each region.


```python
# Visualizing the number of countries per region
countries_per_region.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Countries per Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45, ha="right")
plt.show()
```

This simple example illustrates how Pandas can be used to load and preprocess data, NumPy for data manipulation, and Matplotlib for data visualization, all of which are essential steps in a deep learning workflow.

Although we did not build a deep learning model here, these steps lay the groundwork for data preparation and analysis, which are critical before any deep learning algorithms can be applied.

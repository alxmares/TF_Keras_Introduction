# Matplotlib Tutorial: Basics

Welcome to the basics Matplotlib tutorial. In this section, we'll cover the basics of Matplotlib, a comprehensive library for creating static, animated, and interactive visualizations in Python. Let's start with simple plots and customization options.


```python
# Importing Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Line Plot

Let's start with a simple line plot showing a sine wave.


```python
# Creating a sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting the sine wave
plt.plot(x, y)
plt.title('Simple Sine Wave Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## Multiple Plots

You can easily plot multiple lines on the same graph.


```python
# Creating multiple lines
y2 = np.cos(x)

# Plotting multiple lines
plt.plot(x, y, '-r', label='sin(x)')
plt.plot(x, y2, '-b', label='cos(x)')
plt.legend(loc='upper right')
plt.title('Sine and Cosine Wave Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Scatter Plot

Scatter plots are used to display the relationships between variables.


```python
# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Random Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Histogram

Histograms are great for visualizing the distribution of data points.


```python
# Generating random data
data = np.random.randn(1000)

# Creating a histogram
plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Subplots

Subplots allow you to place multiple plots in a figure.


```python
# Creating subplots
fig, ax = plt.subplots(2, 1)

x = np.linspace(0, 2*np.pi, 100)
ax[0].plot(x, np.sin(x), 'r') # row 0
ax[1].plot(x, np.cos(x), 'b') # row 1

ax[0].set_title('Sine Function')
ax[1].set_title('Cosine Function')

# Automatically adjust the subplot layout
plt.tight_layout()
plt.show()
```

## 3D Plot

Creating 3D plots for visualizing data points in three dimensions.


```python
from mpl_toolkits.mplot3d import Axes3D

# Creating 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional line
zline = np.linspace(-2, 2, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')

plt.show()
```

This concludes our Matplotlib tutorials, where we explored from basic to advanced plotting techniques. With Matplotlib, you can create a wide range of plots and visualizations to effectively communicate your data insights.

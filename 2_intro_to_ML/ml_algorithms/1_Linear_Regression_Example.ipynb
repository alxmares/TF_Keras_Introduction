# Linear Regression

Linear Regression is a foundational statistical method used in machine learning for predictive analysis. It establishes a linear relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. The key objective of linear regression is to predict the value of a dependent variable based on the values of independent variables.

## Key Concepts

- **Linear Relationship**: Assumes that there is a linear correlation between the input variables (independent) and the single output variable (dependent).
- **Coefficients**: The weights assigned to the independent variables indicating their relative importance.
- **Intercept**: The point where the regression line crosses the y-axis when all independent variables are zero.
- **Least Squares**: The method used to estimate the coefficients by minimizing the sum of the squares of the differences between observed and predicted values.

## Applications

- Predicting sales based on advertising budget.
- Estimating house prices from their characteristics.
- Forecasting economic trends.

## Advantages

- Simple to implement and understand.
- Provides a clear indication of the relevance of each independent variable.

## Disadvantages

- Assumes a linear relationship, which may not always hold true.
- Sensitive to outliers.

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

## Load data and split

```python
# Load the diabetes dataset (for demonstration purposes)
diabetes = datasets.load_diabetes()
X = diabetes.data[:, np.newaxis, 2]  # Use only one feature for simplicity
y = diabetes.target

# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Fit step

```python
# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)
```

## Graph

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)

# Plot outputs
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
